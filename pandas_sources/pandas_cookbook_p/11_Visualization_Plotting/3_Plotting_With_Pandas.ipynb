```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
index = ['Atiya', 'Abbas', 'Cornelia', 'Stephanie', 'Montie']
```

```python
data = {'Apples' : [20, 10, 40, 20, 50], 'Oranges' : [35, 40, 25, 19, 33]}
```

```python
df = pd.DataFrame(index = index, data=data)
df
```

### Bar plot
### index - as labels (x - axis), column values - as bar height (y - axis)

```python
color = ['0.2', '0.7']
fig_size = (16, 4)
```

```python
df.plot(kind='bar', color = color, figsize=fig_size)
```

### KDE 
### x-axis - each column values, y-axis - Probability density of the values

```python
df.plot(kind='kde', color=color, figsize=fig_size)
```

### combining all plots in same fig 

```python
fig, axes = plt.subplots(1, 3, figsize=fig_size)
ax1, ax2, ax3 = axes

fig.suptitle('Two Variable Plots', size=20, y=1.02)

# line plot (x-axes - index, y-axes - values)
df.plot(kind='line', color=color, ax=ax1, title='Line plot')

# scatter plot (columns as x and y axes , points as values, no index used)
df.plot(x='Apples', y='Oranges', kind='scatter', ax=ax2, title='Scatterplot')

# bar plot (xa-xes - index, y-axes - columns values ( as bar height))
df.plot(kind='bar', color=color, ax=ax3, title='Bar plot')
```

### putting all one variable plots in single figure

```python
fig, axes = plt.subplots(1, 3, figsize=fig_size)
ax1, ax2, ax3 = axes

fig.suptitle('One Variable Plots', size=20, y=1.02)

# kde plot - columns values as x- axis, pdf as y-axis
df.plot(kind='kde', color=color, ax=ax1, title='KDE plot')

# box plot - column names as x-axis, values as y-axis for boxing
df.plot(kind='box', ax=ax2, title='Box plot')

# histogram - columns values as x-axis, bar height as refrequncy of the values
df.plot(kind='hist', color=color, ax=ax3, title='Histogram')
```

### plotting by specifying the pandas columns

```python
fig, axes = plt.subplots(1, 3, figsize=fig_size)
ax1, ax2, ax3 = axes

df_sorted = df.sort_values('Apples')
df_sorted.plot(x='Apples', y='Oranges', kind='line', ax=ax1)

df.plot(x='Apples', y='Oranges', kind='bar', ax=ax2)

df.plot(x='Apples', kind='kde', ax=ax3)
```
