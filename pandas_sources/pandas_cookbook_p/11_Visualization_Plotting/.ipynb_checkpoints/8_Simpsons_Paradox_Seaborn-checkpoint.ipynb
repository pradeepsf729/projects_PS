```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.cm import Greys
```

```python
diamonds = pd.read_csv('data/diamonds.csv')
diamonds.tail(10)
```

### creating categorical columns

```python
diamonds[['cut', 'color', 'clarity']].dtypes
```

```python
cut_cats = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
color_cats = ['J', 'I', 'H', 'G', 'F', 'E', 'D']
clarity_cats = ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']
```

```python
diamonds['cut'] = pd.Categorical(diamonds['cut'], categories=cut_cats, ordered=True)
```

```python
diamonds['color'] = pd.Categorical(diamonds['color'], categories=color_cats, ordered=True)
```

```python
diamonds['clarity'] = pd.Categorical(diamonds['clarity'], categories=clarity_cats, ordered=True)
```

```python
diamonds[['cut', 'color', 'clarity']].dtypes
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(14,4))

sns.barplot(x='color', y='price', data=diamonds, ax=ax1)
sns.barplot(x='cut', y='price', data=diamonds, ax=ax2)
sns.barplot(x='clarity', y='price', data=diamonds, ax=ax3)

fig.suptitle('Price Decreasing with Increasing Quality?') 
```

#### From above chart
#### There seems to be a decreasing trend for color and price. 
#### The highest quality diamonds with cut (Ideal) and clarity levels (IF) also have low prices. 

```python
sns.catplot(x='color', y='price',
               col='clarity', col_wrap=4,
               data=diamonds, kind='bar')
```

#### From the above plot
#### Although price appears to decrease as the quality of color increases, 
#### but it does not do so when clarity is at its highest level (IF), last bar plot.
#### for highest clarify the prices increases as color increases.

#### plotting similar plots, but with the dismond size (carat)

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(14,4))

sns.barplot(x='color', y='carat', data=diamonds, ax=ax1)
sns.barplot(x='cut', y='carat', data=diamonds, ax=ax2)
sns.barplot(x='clarity', y='carat', data=diamonds, ax=ax3)

fig.suptitle('Diamond size decreases with quality') 
```

#### the above chart reveals that, the better the quality, the size of the diamond decreases.

#### Categorize the carat size

```python
diamonds['carat_category'] = pd.qcut(diamonds.carat, 5)
```

```python
diamonds['carat_category'].value_counts()
```

```python
from matplotlib.cm import Greys
greys = Greys(np.arange(50,250,40))
```

```python
g = sns.catplot(x='clarity', y='price',
                   data=diamonds, hue='carat_category',
                   col='color', col_wrap=4,
                   kind='point', palette=greys)

g.fig.suptitle('Diamond price by size, color and clarity',y=1.02, size=20)
```

####  The above plot reveals that - 
#### higher quality diamonds in fact cost more money when they are segmented based on size.
#### the earlier plot when plotting without the carat size, was showing that the price has downward trend on color.
#### Adding other components, proved it is not the case, 

#### this confirms that diamonds do indeed become more expensive as their quality increases, as
#### long as we hold the carat size as the constant.

#### this is called the Simpson's paradox.

#### Simpson's paradox occurs when one group shows a higher result than another group, when all the
#### data is aggregated, but it shows the opposite when the data is subdivided into different segments.

#### Plotting the similar bivariate relation ship across variables using Seaborn PairGrid

```python
g = sns.PairGrid(diamonds, height=5, 
                 x_vars=["color", "cut", "clarity"], y_vars=["price"])

g.map(sns.barplot)

g.fig.suptitle('Replication of Category vs Price relationship with PairGrid', y=1.02)
```
