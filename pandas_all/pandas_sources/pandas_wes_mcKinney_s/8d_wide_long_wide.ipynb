### 1. Long to Wide

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('examples/macrodata.csv')
data.head()
```

```python
periods = pd.PeriodIndex(year=data.year, quarter=data.quarter,
                         name='date')
```

```python
len(periods)
```

```python
len(data)
```

```python
columns = pd.Index(['realgdp', 'infl', 'unemp'], name='item')
data = data.reindex(columns=columns)
data.index = periods.to_timestamp('D', 'end')
ldata = data.stack().reset_index().rename(columns={0: 'value'})
```

```python
columns
```

```python
data
```

```python
ldata
```

```python
ldata[:10]
```

```python
pivoted = ldata.pivot('date', 'item', 'value')
pivoted[:5]
```

```python
ldata['value2'] = np.random.randn(len(ldata))
ldata[:10]
```

```python
pivoted = ldata.pivot('date', 'item')
pivoted[:5]
```

```python
unstacked = ldata.set_index(['date', 'item'])
unstacked[:7]
```

```python
unstacked = ldata.set_index(['date', 'item']).unstack('item')
unstacked[:7]
```

### 2. Wide to Long

```python
df = pd.DataFrame({'key': ['foo', 'bar', 'baz'],
                   'A': [1, 2, 3],
                   'B': [4, 5, 6],
                   'C': [7, 8, 9]})
df
```

```python
melted = pd.melt(df, ['key'])
melted
```

```python
#melt: It's like bringing the sweet and lid of a cooker onto the sides
```

```python
#pivot is the opposite of melt
```

```python
reshaped = melted.pivot('key', 'variable', 'value')
reshaped
```

```python
reshaped.reset_index()
```

```python
pd.melt(df, id_vars=['key'], value_vars=['A', 'B'])
```

```python
pd.melt(df, value_vars=['A', 'B', 'C'])
```

```python
pd.melt(df, value_vars=['key', 'A', 'B'])
```

```python
pd.melt(df, value_vars=['A'])
```

```python
pd.melt(df, value_vars=['B'])
```

```python
c_melted = pd.melt(df, value_vars=['C'])
```

```python
c_melted
```

```python
c_melted.pivot()
```

```python
c_melted
```

```python
reshaped = c_melted.pivot(c_melted.index, 'variable')
reshaped
```
