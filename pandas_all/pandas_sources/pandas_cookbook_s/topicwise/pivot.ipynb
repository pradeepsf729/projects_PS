```python
stocks = pd.read_csv('prices-split-adjusted.csv',
                     usecols=['date', 'symbol', 'open', 'close'],
                     parse_dates=['date'],
                     index_col='date')
stocks.head()
```

```python
stocks_small = stocks.loc['2016-02-01':'2016-03-01'].reset_index()
stocks_small.head()
```

```python
pivoted = stocks_small.pivot(index='symbol', columns='date')
pivoted
```

```python
pivoted = stocks_small.pivot(index='date', columns='symbol')
pivoted
```

```python
pivoted2 = stocks_small.pivot(index='symbol', columns='date', values='close')
pivoted2
```
