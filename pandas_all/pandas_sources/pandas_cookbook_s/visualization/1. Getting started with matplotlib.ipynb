```python
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd 
import numpy as np
```

* Stateful current plotting thing
* Object oriented: 

```python
x = [-3, 5, 7]
y = [10, 2, 5]
```

# Statefuol method

```python
plt.figure(figsize=(15,3))
plt.plot(x, y)
plt.xlim(0, 10)
plt.ylim(-3, 8)
plt.xlabel('X Axis')
plt.ylabel('Y axis')
plt.title('Line Plot')
plt.suptitle('Figure Title', size=20, y=1.03)
```

# object Oriented

```python
fig, ax = plt.subplots(figsize=(15,3))
ax.plot(x, y)
ax.set_xlim(0, 10)
ax.set_ylim(-3, 8)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_title('Line Plot')
fig.suptitle('Figure Title', size=20, y=1.03)
```
