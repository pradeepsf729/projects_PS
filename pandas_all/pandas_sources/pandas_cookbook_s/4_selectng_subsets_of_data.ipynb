### 5. Speeding up scalar selection

```python
df_college = pd.read_csv('data/college.csv', index_col='INSTNM')
college_name = 'Texas A & M University-College Station'
df_college.loc[college_name, 'UGDS_WHITE']
```

```python
df_college.at[college_name, 'UGDS_WHITE']
```

```python
%timeit df_college.loc[college_name, 'UGDS_WHITE']
```

```python
%timeit df_college.at[college_name, 'UGDS_WHITE']
```

```python
row_num = df_college.index.get_loc(college_name)
col_num = df_college.columns.get_loc('UGDS_WHITE')
row_num, col_num
```

```python
%timeit df_college.iloc[row_num, col_num]
```

```python
%timeit df_college.iat[row_num, col_num]
```

### Notes: 

.at == .loc both need strings or keys \
.iat == .iloc both need integers as keys

.at amd .iat are fatser comapred to .loc and .iloc

```python
state.iat[1000]
```

### 7. Slicing lexicographically

```python
df_college = df_college.sort_index()
```

```python
df_college.loc['Sp':'Su'].head()
```

```python
#when the index is sortred, .loc works based on lexicographic index
```

```python
df_college.loc['A':'B'].sample(3)
```

```python
len(df_college.loc['A':'B'])
# this contains the rows with B as well, the last index will not be ignored
# this is a special case
```

```python
df_college.index.is_monotonic_decreasing
```

```python
df_college = df_college.sort_index(ascending=False)
df_college.index.is_monotonic_decreasing
```
