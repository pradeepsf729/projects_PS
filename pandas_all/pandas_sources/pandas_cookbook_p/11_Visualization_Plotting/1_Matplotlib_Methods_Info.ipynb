```python
import matplotlib.pyplot as plt
```

### Using stateful approach, (all attributes are sent to plt object and modified)

```python
x = [-3, 5, 7]
y = [10, 2, 5]
```

```python
plt.figure(figsize=(15, 3))
```

```python
plt.xlim(0, 10)
plt.ylim(-3, 8)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Line plot')
plt.suptitle('Figure Title', size=20, y=1.03)
plt.plot(x, y)
```

### using stateless Object oriented way ( using variable referencing specific axis/figure object, instead of plt)

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(x, y)
ax.set_xlim(0, 10)
ax.set_ylim(-3, 8)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_title('Line plot')
fig.suptitle('Figure Title', size=20, y=1.03)
```

#### spine objects are the border lines around the axes

```python
spines = ax.spines
[x for x in dir(spines) if x[0] != '_']
```

```python
[(k,v) for k,v in spines.items()]
```

```python
spine_left = spines['left']
spine_left.set_position(('outward', - 100))
spine_left.set_linewidth(5)
```

```python
spine_bottom = spines['bottom']
spine_bottom.set_visible(False)
```

```python
fig
```

#### the axes objects represents the two x and y axes, which are part of Axes object.

```python
ax.xaxis.grid(True, which='major', linewidth=2, color='black', linestyle='--')
ax.xaxis.set_ticks([.2, .4, .55, .93])
ax.xaxis.set_label_text('X Axis', family='Verdana', fontsize=15)
ax.set_ylabel('Y Axis', family='DejaVu Sans', fontsize=20)
ax.set_yticks([.1, .9])
ax.set_yticklabels(['point 1', 'point 9'], rotation=45)
fig
```
