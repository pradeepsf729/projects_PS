```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
employee = pd.read_csv('data/employee.csv', parse_dates=['HIRE_DATE', 'JOB_DATE'])
```

```python
employee.head(4)
```

```python
sns.countplot(y='DEPARTMENT', data=employee)  # in seaborn, caluculation done as part of plot API
```

#### Similar above , plot with pandas

```python
employee['DEPARTMENT'].value_counts().plot(kind='barh')
```

```python
ax = sns.barplot(x='RACE', y='BASE_SALARY', data=employee)
ax.figure.set_size_inches(16, 4)
```

### similar, avg plot with pandas

```python
avg_sal = employee.groupby('RACE', sort=False)['BASE_SALARY'].mean()
```

```python
ax = avg_sal.plot(kind='bar', rot=0, figsize=(16,4), width=.8)

ax.set_xlim(-.5, 5.5)

ax.set_ylabel('Mean Salary')
```

### automatic grouping with legend from the barplot API itself fro seaborn

```python
ax = sns.barplot(x='RACE', y='BASE_SALARY', hue='GENDER', data=employee)
ax.figure.set_size_inches(16,4)
```

### Creating the similar above plot with pandas

```python
emp_sal_grp = employee.groupby(['RACE', 'GENDER'])['BASE_SALARY'].mean()
```

```python
emp_sal_grp
```

```python
emp_sal_grp = emp_sal_grp.unstack('GENDER')
```

```python
emp_sal_grp
```

```python
emp_sal_grp.plot(kind='bar', figsize=(16, 4), rot=0, width=0.8)
```

### Creating boxplot using seaborn

```python
ax = sns.boxplot(x='GENDER', y='BASE_SALARY',
                data=employee,
                hue='RACE', palette='Greys')
ax.figure.set_size_inches(14,4)
```

### Creating similar box plot with pandas

```python
fig, ax_array = plt.subplots(1, 2, figsize=(14,4), sharey=True)

ax1, ax2 = ax_array
g0, g1 = ['Female', 'Male']

fm = employee.query('GENDER == @g0')
fm.boxplot(by='RACE', column='BASE_SALARY', ax=ax1, rot=20)
ax1.set_title(g0 + ' Salary')
ax1.set_xlabel('')


ma = employee.query('GENDER == @g1')
ma.boxplot(by='RACE', column='BASE_SALARY', ax=ax2, rot=20)
ax2.set_title(g1 + ' Salary')
ax2.set_xlabel('')

fig.suptitle('')
```
