```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
employee = pd.read_csv('data/employee.csv', parse_dates=['HIRE_DATE', 'JOB_DATE'])
```

```python
employee.head(4)
```

#### calculate year of experience from the hire date 

```python
days_to_hire = pd.to_datetime('21-1-2016') - employee['HIRE_DATE']
```

```python
one_year_unit = pd.Timedelta('365 days')
```

```python
employee['YEARS_EXPERIENCE'] = days_to_hire / one_year_unit
```

```python
employee[['HIRE_DATE', 'YEARS_EXPERIENCE']].head(4)
```

```python
ax = sns.regplot(x='YEARS_EXPERIENCE', y='BASE_SALARY', data=employee)
ax.figure.set_size_inches(14, 4)
```

### plotting with using third variable to the scatter plot
### with regressions calculated

```python
grid = sns.lmplot(x='YEARS_EXPERIENCE', y='BASE_SALARY',
                    hue='GENDER', palette='Greys',
                    scatter_kws={'s':10}, data=employee)
grid.fig.set_size_inches(14, 4)
```

```python
type(grid)
```

### adding more variables 

```python
grid2 = sns.lmplot(x='YEARS_EXPERIENCE', y='BASE_SALARY',
                    hue='GENDER', col='RACE', col_wrap=3,
                    palette='Greys', sharex=False,
                    line_kws = {'linewidth':5},
                    data=employee)

grid2.set(ylim=(20000, 120000))
```

```python
type(grid2)
```

#### for each type of columns value and grid plot will be created

```python
grid2 = sns.lmplot(x='YEARS_EXPERIENCE', y='BASE_SALARY',
                    hue='GENDER', col='DEPARTMENT', col_wrap=8,
                    palette='Greys', sharex=False,
                    line_kws = {'linewidth':5},
                    data=employee)

grid2.set(ylim=(20000, 120000))
```

### creating the similar plot for other categorical columns

#### reducing number of categories

```python
deps = employee['DEPARTMENT'].value_counts().index[:2]
races = employee['RACE'].value_counts().index[:3]
is_dep = employee['DEPARTMENT'].isin(deps)
is_race = employee['RACE'].isin(races)
```

```python
emp2 = employee[is_dep & is_race].copy()
```

```python
emp2['DEPARTMENT'] = emp2['DEPARTMENT'].str.extract('(HPD|HFD)',
                                                        expand=True)
```

```python
emp2[['RACE', 'DEPARTMENT']].head(4)
```

```python
emp2['DEPARTMENT'].value_counts()
```

```python
emp2['RACE'].value_counts()
```

#### plotting what are most common departs based on years of exp

```python
common_depts = employee.groupby('DEPARTMENT') \
                .filter(lambda x: len(x) > 50)
```

```python
ax = sns.violinplot(x='YEARS_EXPERIENCE',
                    y='GENDER',
                    data=common_depts)
ax.figure.set_size_inches(10,4)
```

```python
sns.catplot(x='YEARS_EXPERIENCE', y='GENDER',
                col='RACE',row='DEPARTMENT',
                height=3, aspect=2,
                data=emp2, kind='violin')
```
